## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Top Spenders
# Group by SN
sn_group = purchase_df.groupby("SN")

# Total items purchased
purch_count = sn_group["Purchase ID"].count()

# Total spent
avg_purch_price = sn_group["Price"].sum()

# Merge
SN_merge = pd.merge(purch_count,avg_purch_price, on="SN", how="outer")

# Sort by price
SN_merge = SN_merge.sort_values(["Price"], ascending=False)

# Rename columns
SN_merge = SN_merge.rename(columns = {"Purchase ID": "Purchase Count", "Price": "Total Purchase Value"})

# Add column - Avg purchase price
SN_merge["Average Purchase Price"] = SN_merge["Total Purchase Value"]/SN_merge["Purchase Count"]

# Format
SN_merge["Average Purchase Price"] = SN_merge["Average Purchase Price"].map('${:,.2f}'.format)
SN_merge["Total Purchase Value"] = SN_merge["Total Purchase Value"].map('${:,.2f}'.format)

# Top 5 Spenders
top_5_spenders = SN_merge.head(5)
top_5_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Top Items purchased
# Group by Item ID
Item_group = purchase_df.groupby("Item ID")

# Total items purchased
item_purch_count = Item_group["Purchase ID"].count()

# Total spent
item_avg_purch_price = Item_group["Price"].sum()

# Merge
item_merge = pd.merge(item_purch_count,item_avg_purch_price, on="Item ID", how="outer")

# Sort by frequency
item_merge = item_merge.sort_values(["Purchase ID"], ascending=False)

# Rename columns
item_merge = item_merge.rename(columns = {"Purchase ID": "Purchase Count", "Price": "Total Purchase Value"})

# Add column - Item Price
item_merge["Item Price"] = item_merge["Total Purchase Value"]/item_merge["Purchase Count"]
item_name = purchase_df.loc[:, ("Item ID", "Item Name")]
item_name_merge = pd.merge(item_merge, item_name, on="Item ID", how="left")
all_items = item_name_merge.drop_duplicates(subset=['Item ID'])
all_items = all_items.set_index("Item ID")

# Rearrange column names
# Source: https://www.kite.com/python/answers/how-to-reorder-columns-in-a-pandas-dataframe-in-python
cols = ["Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]
all_items = all_items.reindex(columns = cols)
all_items

# Format
all_items["Item Price"] = all_items["Item Price"].map('${:,.2f}'.format)
all_items["Total Purchase Value"] = all_items["Total Purchase Value"].map('${:,.2f}'.format)

# Top 5 Spenders
top_5_items = all_items.head(5)
top_5_items
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
# # Most profitable items

# Group by Item ID
Item_group = purchase_df.groupby("Item ID")

# Total items purchased
item_purch_count = Item_group["Purchase ID"].count()

# Total spent
item_avg_purch_price = Item_group["Price"].sum()

# Merge
item_merge = pd.merge(item_purch_count,item_avg_purch_price, on="Item ID", how="outer")

# Sort by total profit
item_merge = item_merge.sort_values(["Price"], ascending=False)
item_merge

#Rename columns
item_merge = item_merge.rename(columns = {"Purchase ID": "Purchase Count", "Price": "Total Purchase Value"})

# Add column - Item Price
item_merge["Item Price"] = item_merge["Total Purchase Value"]/item_merge["Purchase Count"]
item_name = purchase_df.loc[:, ("Item ID", "Item Name")]
item_name_merge = pd.merge(item_merge, item_name, on="Item ID", how="left")
all_items = item_name_merge.drop_duplicates(subset=['Item ID'])
all_items = all_items.set_index("Item ID")

# Rearrange column names
# Source: https://www.kite.com/python/answers/how-to-reorder-columns-in-a-pandas-dataframe-in-python
cols = ["Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]
all_items = all_items.reindex(columns = cols)
all_items

# Format
all_items["Item Price"] = all_items["Item Price"].map('${:,.2f}'.format)
all_items["Total Purchase Value"] = all_items["Total Purchase Value"].map('${:,.2f}'.format)

# Top 5 Spenders
top_5_items = all_items.head(5)
top_5_items
```
